# Logistic Regression for Binary Classification
In logistic regression for binary classification, the probability the positive class (we encode positive class as '1' and negative class as '0') given an input example $\mathbf{x}$ is modeled using a <em>logistic function</em> as given below
$$\hat{p}(\mathbf{x})=\hat{P}(y=1|\mathbf{x}) = \frac{1}{1+e^{-(\mathbf{w}^T\mathbf{x}+w_0)}} = \frac{e^{\mathbf{w}^T\mathbf{x}+w_0}}{e^{\mathbf{w}^T\mathbf{x}+w_0}+1}$$
where $\mathbf{w}$ and $w_0$ are the model coefficients and intercept. Thus, the probability of negative class (i.e. class 0) given input $\mathbf{x}$ can be given as
$$\hat{P}(y=0|\mathbf{x}) = 1- \hat{p}(\mathbf{x}) = 1-\frac{1}{1+e^{-(\mathbf{w}^T\mathbf{x}+w_0)}}=\frac{1}{1+e^{\mathbf{w}^T\mathbf{x}+w_0}}$$
The parameters $\mathbf{w}$ and $w_0$ are found by solving the following optimization problem
$$\underset{\mathbf{w}, w_0}{\text{min}}\quad C\sum_{i=1}^n -y_i\text{log }\hat{p}(\mathbf{x}_i)-(1-y_i)\text{log }(1-\hat{p}(\mathbf{x}_i))+r(\mathbf{w})$$
where $r(\mathbf{w})$ denotes the penalty term on the coefficients $\mathbf{w}$. For $l_2$ penalty $r(\mathbf{w})=\frac{1}{2}||\mathbf{w}||^2_2$ and for $l_1$ penalty, $r(\mathbf{w})=||\mathbf{w}||_1$.

Scikit-learn provides the class <em>sklearn.linear_model.LogisticRegression</em> for Logistic regression. A complete documentation of the same can be found [here](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression).

# Logistic Regression for Multiclass Classification
Let there be $K$ classes ($K > 2$) and they are numbered from $1$ to $K$. Thus, in this case, each label $y_i$ can be one of $\{1, 2, \cdots, K\}$. In logistic regression for $K$ classes where $K>2$ (also known as <em>multinomial logistic regression</em>), $K$ coefficent vectors $\mathbf{w}_1, \mathbf{w}_2, \cdots, \mathbf{w}_K$ and $K$ intercepts $w_{0,1}, w_{0,2}, \cdots, w_{0,K}$ are used as the parameters of the classifier. The probability of the $k$-th class given an input example $\mathbf{x}$ is modelled using a <em>softmax function</em> as given below:
$$\hat{p}_k(\mathbf{x})=\hat{P}(y=k|\mathbf{x})=\frac{e^{\mathbf{w}_k+w_{0,k}}}{\sum_{j=1}^K e^{\mathbf{w}_j+w_{0,k}}}$$
The parameters of the model are found by solving the following optimization problem
$$\underset{(\mathbf{w}_k, w_{0,k})_{k=1}^K}{\text{min}}\quad -C\sum_{i=1}^n\sum_{k=1}^K[y_i=k]\text{log }\hat{p}_k(\mathbf{x}) + r(\mathbf{w}_1, \cdots, \mathbf{w}_K)$$
where $[P]$ dontes the indicator variable which equates to $1$ if $P$ is true and to $0$ otherwise and $r(\mathbf{w}_1, \cdots, \mathbf{w}_K)$ is the penalty term. In Scikit-learn, you can use the <em>sklearn.linear_model.LogisticRegression</em> class (same as binary logistic regression) for multinomial logistic regression. The complete documentation can be found [here](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression).

### Decision Boundary for binary classification

```python
X, y = make_classification(200, 2, n_informative=2, n_redundant=0, weights=[.5, .5], random_state=15)
clf = LogisticRegression().fit(X[:100], y[:100])
```

```python
xx, yy = np.mgrid[-5:5:.01, -5:5:.01]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = clf.predict_proba(grid)[:, 1].reshape(xx.shape)
```

```python
f, ax = plt.subplots(figsize=(8, 6))
ax.contour(xx, yy, probs, levels=[.5], cmap="Greys", vmin=0, vmax=.6)

ax.scatter(X[100:,0], X[100:, 1], c=y[100:], s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)

ax.set(aspect="equal",
       xlim=(-5, 5), ylim=(-5, 5),
       xlabel="$X_1$", ylabel="$X_2$")
```

### Decision boundary with prediction probability

```python
f, ax = plt.subplots(figsize=(8, 6))
contour = ax.contourf(xx, yy, probs, 25, cmap="RdBu",
                      vmin=0, vmax=1)
ax_c = f.colorbar(contour)
ax_c.set_label("$P(y = 1)$")
ax_c.set_ticks([0, .25, .5, .75, 1])

ax.scatter(X[100:,0], X[100:, 1], c=y[100:], s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)

ax.set(aspect="equal",
       xlim=(-5, 5), ylim=(-5, 5),
       xlabel="$X_1$", ylabel="$X_2$")
```

### Decision Boundary for three Class Classification
The original code can be found [here](https://stackoverflow.com/questions/70891146/how-to-plot-descion-boundary-for-three-classes-in-logistic-regression).

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

custom_cmap = ListedColormap(['#b4a7d6','#93c47d','#fff2cc'])

iris = load_iris()

x_index = 0
y_index = 1
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
```

```python
lr = LogisticRegression(multi_class='multinomial', random_state=42, max_iter=500)
lr.fit(iris.data[:, [0, 1]], iris.target)
```

```python
x0, x1 = np.meshgrid(
    np.linspace(iris.data[:, x_index].min(), iris.data[:, x_index].max(), 500).reshape(-1, 1),
    np.linspace(iris.data[:, y_index].min(), iris.data[:, y_index].max(), 500).reshape(-1, 1)
)

X_new = np.c_[x0.ravel(), x1.ravel()]
y_pred = lr.predict(X_new)
zz = y_pred.reshape(x0.shape)

plt.figure(figsize=(10, 5))
plt.contourf(x0, x1, zz, cmap=custom_cmap)
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)

plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])

plt.tight_layout()
plt.show()
```

## Logistic Function: Logistic Regression vs. Linear Regression for Binary Classification
The original code for the following example can be found [here](https://scikit-learn.org/stable/auto_examples/linear_model/plot_logistic.html#sphx-glr-auto-examples-linear-model-plot-logistic-py).

```python
# Code source: Gael Varoquaux
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression, LinearRegression
from scipy.special import expit

# Generate a toy dataset, it's just a straight line with some Gaussian noise:
xmin, xmax = -5, 5
n_samples = 100
np.random.seed(0)
X = np.random.normal(size=n_samples)
y = (X > 0).astype(float)
X[X > 0] *= 8
X += 0.3 * np.random.normal(size=n_samples)

X = X[:, np.newaxis]

# Fit the classifier
clf = LogisticRegression(C=1e5)
clf.fit(X, y)

# and plot the result
plt.figure(1, figsize=(5, 3))
plt.clf()
plt.scatter(X.ravel(), y, color="black", zorder=20)
X_test = np.linspace(-10, 20, 300)

loss = expit(X_test * clf.coef_ + clf.intercept_).ravel()
plt.plot(X_test, loss, color="red", linewidth=3)

ols = LinearRegression()
ols.fit(X, y)
plt.plot(X_test, ols.coef_ * X_test + ols.intercept_, linewidth=1)
plt.axhline(0.5, color=".5")

plt.ylabel("y")
plt.xlabel("X")
plt.xticks(range(-10, 20, 2))
plt.yticks([0, 0.5, 1])
plt.ylim(-0.25, 1.25)
plt.xlim(-10, 15)
plt.legend(
    ("Logistic Regression Model", "Linear Regression Model"),
    loc="lower right",
    fontsize="small",
)
plt.tight_layout()
plt.show()
```

# Linear Discriminant Analysis
Linear Discriminant Analysis (LDA) models the class conditional desnity i.e. the probability of the features given a class (denoted as $P(\mathbf{x}|y=k)$, for $k=1, 2, \ldots, K$) using a multivariate Gaussian distribution. Thus, we can write
$$P(\mathbf{x}|y=k) = \frac{1}{(2\pi)^{d/2}|\Sigma_k|^{1/2}}\text{exp} \left(-\frac{1}{2}(\mathbf{x}-\mu_k)^T\Sigma^{-1}(\mathbf{x}-\mu_k)\right)$$
where $\mu_k$ and $\Sigma_k$ are the mean vector and the covariance matrix of the multivariate Gaussian distribution for the $k$-th class and $d$ is the dimension of the feature vector $\mathbf{x}$. Additional, in LDA, we assume that the covariance matrix for all classes is same i.e. $\Sigma_k=\Sigma$ for all $k=1, 2, \dots, K$. The prediction probabilities are obtained using Bayes rule
$$P(y=k|\mathbf{x})=\frac{P(\mathbf{x}|y=k)P(y=k)}{P(\mathbf{x})}=\frac{P(\mathbf{x}|y=k)P(y=k)}{\sum_{j=1}^KP(\mathbf{x}|y=j)P(y=j)}$$
We select the class for which $P(y=k|\mathbf{x})$ is maximum as the predicted class. In Scikit-learn, the class <em>sklearn.discriminant_analysis.LinearDiscriminantAnalysis</em> is provided for performing LDA. The complete documentation of the same can be found [here](https://scikit-learn.org/stable/modules/lda_qda.html#dimensionality-reduction-using-linear-discriminant-analysis).

## Decision Boundary of LDA
The following code snippets are the modified version of [this code](https://scikit-learn.org/stable/auto_examples/classification/plot_lda_qda.html).

### Datasets generation functions

```python
import numpy as np


def dataset_fixed_cov():
    """Generate 2 Gaussians samples with the same covariance matrix"""
    n, dim = 300, 2
    np.random.seed(0)
    C = np.array([[1, 0.0], [0.0, 0.4]])
    X = np.r_[
        np.dot(np.random.randn(n, dim), C),
        np.dot(np.random.randn(n, dim), C) + np.array([0, 1]),
        np.dot(np.random.randn(n, dim), C) + np.array([1.5, 0.5]),
    ]
    y = np.hstack((np.zeros(n), np.ones(n), 2*np.ones(n)))
    return X, y


def dataset_cov():
    """Generate 2 Gaussians samples with different covariance matrices"""
    n, dim = 300, 2
    np.random.seed(0)
    C1 = np.array([[1, 0.0], [0.0, 0.4]])
    C2 = np.array([[0.6, 0.0], [0.0, 0.6]])
    X = np.r_[
        np.dot(np.random.randn(n, dim), C1),
        np.dot(np.random.randn(n, dim), C1) + np.array([0, 1]),
        np.dot(np.random.randn(n, dim), C2) + np.array([1.5, 0.5]),
    ]
    y = np.hstack((np.zeros(n), np.ones(n), 2*np.ones(n)))
    return X, y
```

### Plot functions

```python
from scipy import linalg
import matplotlib as mpl
from matplotlib.colors import ListedColormap

custom_cmap = ListedColormap(['#ffb0b0','#b0b0ff', '#b0ffb0'])

def plot_data(lda, X, y, y_pred, fig_index):
    splot = plt.subplot(1, 2, fig_index)
    if fig_index == 1:
        plt.title("Linear Discriminant Analysis")
        plt.ylabel("Data with\n fixed covariance")
    elif fig_index == 2:
        plt.title("Quadratic Discriminant Analysis")
    elif fig_index == 3:
        plt.ylabel("Data with\n varying covariances")

    X0, X1, X2 = X[y == 0], X[y == 1], X[y == 2]

    # class 0, 1 and 2: areas
    nx, ny = 200, 100
    x_min, x_max = X[:, 0].min(), X[:, 0].max()
    y_min, y_max = X[:, 1].min(), X[:, 1].max()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
    Z = lda.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=custom_cmap)

    # class 0: dots
    plt.scatter(X0[:, 0], X0[:, 1], marker=".", color="red")

    # class 1: dots
    plt.scatter(X1[:, 0], X1[:, 1], marker=".", color="blue")

    # class 2: dots
    plt.scatter(X2[:, 0], X2[:, 1], marker=".", color="green")


    # means
    plt.plot(
        lda.means_[0][0],
        lda.means_[0][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )
    plt.plot(
        lda.means_[1][0],
        lda.means_[1][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )

    plt.plot(
        lda.means_[2][0],
        lda.means_[2][1],
        "*",
        color="yellow",
        markersize=15,
        markeredgecolor="grey",
    )

    return splot

def plot_ellipse(splot, mean, cov, color):
    v, w = linalg.eigh(cov)
    u = w[0] / linalg.norm(w[0])
    angle = np.arctan(u[1] / u[0])
    angle = 180 * angle / np.pi  # convert to degrees
    # filled Gaussian at 2 standard deviation
    ell = mpl.patches.Ellipse(
        mean,
        2 * v[0] ** 0.5,
        2 * v[1] ** 0.5,
        180 + angle,
        facecolor=color,
        edgecolor="black",
        linewidth=2,
    )
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.2)
    splot.add_artist(ell)
    splot.set_xticks(())
    splot.set_yticks(())


def plot_lda_cov(lda, splot):
    plot_ellipse(splot, lda.means_[0], lda.covariance_, "red")
    plot_ellipse(splot, lda.means_[1], lda.covariance_, "blue")
    plot_ellipse(splot, lda.means_[2], lda.covariance_, "green")


def plot_qda_cov(qda, splot):
    plot_ellipse(splot, qda.means_[0], qda.covariance_[0], "red")
    plot_ellipse(splot, qda.means_[1], qda.covariance_[1], "blue")
    plot_ellipse(splot, qda.means_[2], qda.covariance_[2], "green")
```

### Plot of decision boundary when the classes share the same covariance matrix

```python
plt.figure(figsize=(10, 8), facecolor="white")
plt.suptitle(
    "Linear Discriminant Analysis vs Quadratic Discriminant Analysis",
    y=0.98,
    fontsize=15,
)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

X, y = dataset_fixed_cov()

# Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(solver="svd", store_covariance=True)
y_pred = lda.fit(X, y).predict(X)
splot = plot_data(lda, X, y, y_pred, fig_index=1)
plot_lda_cov(lda, splot)
#plt.axis("tight")
splot.set_aspect(1)

# Quadratic Discriminant Analysis
qda = QuadraticDiscriminantAnalysis(store_covariance=True)
y_pred = qda.fit(X, y).predict(X)
splot = plot_data(qda, X, y, y_pred, fig_index=2)
plot_qda_cov(qda, splot)
#plt.axis("tight")
splot.set_aspect(1)

plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```

### Plot of decision boundary when the classes have different covariance matrices

```python
plt.figure(figsize=(10, 8), facecolor="white")
plt.suptitle(
    "Linear Discriminant Analysis vs Quadratic Discriminant Analysis",
    y=0.98,
    fontsize=15,
)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

X, y = dataset_cov()

# Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(solver="svd", store_covariance=True)
y_pred = lda.fit(X, y).predict(X)
splot = plot_data(lda, X, y, y_pred, fig_index=1)
plot_lda_cov(lda, splot)
#plt.axis("tight")
splot.set_aspect(1)

# Quadratic Discriminant Analysis
qda = QuadraticDiscriminantAnalysis(store_covariance=True)
y_pred = qda.fit(X, y).predict(X)
splot = plot_data(qda, X, y, y_pred, fig_index=2)
plot_qda_cov(qda, splot)
#plt.axis("tight")
splot.set_aspect(1)

plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
